```python
import os
import shutil

# Define the input and output directories
input_folder = "/content/drive/MyDrive/sample_data/Labels/aachen"
output_folder = "/content/drive/MyDrive/sample_data/Labels/aachen-mask1"

# Create the output folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# List all files in the input folder
files = os.listdir(input_folder)

# Filter out only the files ending with '.png'
mask1 = [file for file in files if file.endswith('_gtFine_color.png')]

# Copy or move the png files to the output folder
for file in mask1:
    file_path = os.path.join(input_folder, file)
    shutil.copy(file_path, output_folder)  # Use shutil.move if you want to move instead of copy
```

```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Set the directory paths for images and masks
img_dir = '/content/drive/MyDrive/sample_data/Images/aachen'
mask_dir = '/content/drive/MyDrive/sample_data/Labels/aachen-mask1'

# Get the list of image files in the directory
img_files = os.listdir(img_dir)

# Display images along with their corresponding masks
for img_file in img_files:
    # Load the image
    img_path = os.path.join(img_dir, img_file)
    img = mpimg.imread(img_path)

    # Construct the corresponding mask filename
    mask_file = img_file.replace('_leftImg8bit.png', '_gtFine_color.png')
    mask_path = os.path.join(mask_dir, mask_file)

    # Load the mask
    mask = mpimg.imread(mask_path)

    # Plot the image and mask side by side
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))
    axes[0].imshow(img)
    axes[0].set_title('Image')
    axes[0].axis('off')
    axes[1].imshow(mask)
    axes[1].set_title('Mask')
    axes[1].axis('off')
    plt.show()
```

```python
import cv2
import numpy as np
import os

def count_unique_colors(image_path):
    # Read the image mask
    image = cv2.imread(image_path)

    # Convert the image to RGB color space
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Flatten the image into a 2D array
    pixels = image_rgb.reshape(-1, 3)

    # Find unique colors and their counts
    unique_colors, counts = np.unique(pixels, axis=0, return_counts=True)

    # Count the number of unique colors
    num_colors = len(unique_colors)

    return num_colors

# Directory containing image masks
mask_dir = '/content/drive/MyDrive/sample_data/Labels/aachen-mask1'

# List of image paths
image_paths =  os.listdir(mask_dir)

# Iterate over each image and count unique colors
for image_name in image_paths:
    image_path = os.path.join(mask_dir, image_name)
    num_colors = count_unique_colors(image_path)
    print(f"Number of unique colors in {image_name}: {num_colors}")
```
